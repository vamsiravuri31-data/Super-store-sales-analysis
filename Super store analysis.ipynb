```python
# Convert Date datatypes
```

```python
df['order_date']=pd.to_datetime(df['order_date'], dayfirst=True)
```

```python
df.head(2)
```

```python
df['ship_date']=pd.to_datetime(df['ship_date'], dayfirst=True)
```

```python
# Fix sales column values, remove commas and extra symbols
```

```python
df['sales']=df['sales'].replace('[$/,]','',regex=True).astype(float)
```

```python
# Create New Columns
```

```python
# Delivery time - Shipping time analysis
```

```python
df['Delivery_days']=(df['ship_date']-df['order_date']).dt.days
```

```python
df.head(2)
```

```python
# Profit ratio
```

```python
df['profit_ratio']=df['profit']/df['sales']*100
```

```python
df.head(2)
```

```python
# Extract order month and year
```

```python
df['order_month']=df['order_date'].dt.month
df['order_year']=df['order_date'].dt.year
```

```python
df.head(2)
```

```python
# Delete unnecessary columns
```

```python
df.drop([
    'order_id',
    'customer_name',
    'product_id',
    'market',
    'order_priority'
], axis=1, inplace=True)
```

```python
df.groupby('ship_mode').agg({
    'Delivery_days': 'mean',
    'ship_mode': 'count'
})
```

```python
df.groupby('discount').agg({
    'sales': 'sum',
    'profit': 'mean',
    'discount': 'count'
})
```

```python
# INSIGHTS
```

```python
# Highest orders in category
high_order_category=df['category'].value_counts().idxmax()
print('Highest orders in category is:', high_order_category)

# Highest orders in sub_category
high_order_sub_category=df['sub_category'].value_counts().idxmax()
print('Highest orders in sub-category is:', high_order_sub_category)

# top region with high orders
top_region=df['region'].value_counts().idxmax()
print('Top region with high orders is:', top_region)

# top segment with high orders
top_segment=df['segment'].value_counts().idxmax()
print('Top segment with high orders is:', top_segment)

# mostly used ship mode to deliver
top_ship_mode=df['ship_mode'].value_counts().idxmax()
print('Top most used ship mode is:', top_ship_mode)

# Peak business month with high orders
top_month=df['order_month'].value_counts().idxmax()
print('Peak business month with high orders is:', top_month)

# Highest sales category
high_sale_category = df.groupby('category')['sales'].sum().idxmax()
print('Highest sales category is:', high_sale_category)

# Highest profits category
high_profit_category = df.groupby('category')['profit'].sum().idxmax()
print('Highest profits category is:', high_profit_category)

# subcategory wise highest sales, profits, orders
high_sub_category_sales_profits_orders = df.groupby('sub_category').agg({'sales':'sum','profit':'sum','sub_category':'count'}).idxmax()
print('Highest sub-category wise sales profits orders are:', high_sub_category_sales_profits_orders)

# Region wise highest profits , sales, orders
high_region_sales_profits_orders = df.groupby('region').agg({'sales':'sum','profit':'sum','region':'count'}).idxmax()
print('Highest Region wise sales profits orders are:', high_region_sales_profits_orders)

# Country wise highest profit, sales, orders
high_country_sales_profits_orders = df.groupby('country').agg({'sales':'sum','profit':'sum','country':'count'}).idxmax()
print('Highest country wise sales profits orders are:', high_country_sales_profits_orders)

# State wise highest sales, profit, orders
high_state_sales_profits_orders = df.groupby('state').agg({'sales':'sum','profit':'sum','state':'count'}).idxmax()
print('Highest state wise sales profits orders are:', high_state_sales_profits_orders)

# Top most month with highest sales, profits,orders
high_month_sales_profits_orders = df.groupby('order_month').agg({'sales':'sum','profit':'sum','order_month':'count'}).idxmax()
print('Top most month with highest sales profits orders are:', high_month_sales_profits_orders)

# Top most year with highest sales, profits,orders
high_year_sales_profits_orders = df.groupby('order_year').agg({'sales':'sum','profit':'sum','order_year':'count'}).idxmax()
print('Top most year with highest sales profits orders are:', high_year_sales_profits_orders)

# Top most year with highest sales, profits,orders
high_product_sales_profits_orders = df.groupby('product_name').agg({'sales':'sum','profit':'sum','product_name':'count'}).idxmax()
print('Top most product with highest sales profits orders are:', high_product_sales_profits_orders)

# Discount given with highest profit
high_discount_sales_profits_orders = df.groupby('discount').agg({'sales':'sum','profit':'mean','discount':'count'}).idxmax()
print('Top most discount with highest sales profits orders are:', high_discount_sales_profits_orders)
```
